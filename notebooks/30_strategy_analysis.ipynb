```python
import pandas as pd

account_df = pd.read_csv('BroadridgeAccount.csv')

inheritance_df = account_df[account_df['SourceofFunds'] == 'Gift/Inheritance']
total_inheritance = len(inheritance_df)
print(f"1. Total Inheritance Accounts: {total_inheritance}")


sweep_dist = inheritance_df['Sweep'].value_counts(normalize=True) * 100
print(f"2. % in Idle Cash (RBC Insured Deposits): {sweep_dist.get('RBC Insured Deposits', 0):.2f}%")


mismatch = inheritance_df[
    (inheritance_df['InvestmentObjectives'].str.contains('Growth|Aggressive', na=False, case=False)) &
    (inheritance_df['RiskTolerance'].isin(['Low Risk', 'Minimal Risk']))
]
print(f"3. Number of 'Risk Mismatch' Accounts: {len(mismatch)}")


edelivery_rate = inheritance_df['EDeliveryStatements'].mean() * 100
print(f"4. E-Delivery Statement Opt-in Rate: {edelivery_rate:.2f}%")


unmanaged_rate = (inheritance_df['AdvisoryAccount'] == False).mean() * 100
print(f"5. Self-Directed (Unmanaged) Rate: {unmanaged_rate:.2f}%")


inheritance_2025 = inheritance_df.copy()
inheritance_2025['OpenDate'] = pd.to_datetime(inheritance_2025['OpenDate'], errors='coerce')
new_2025_count = (inheritance_2025['OpenDate'].dt.year == 2025).sum()
print(f"6. Inheritance Accounts opened in 2025: {new_2025_count}")
```
